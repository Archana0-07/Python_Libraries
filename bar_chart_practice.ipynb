```python
#Programming Languages Popularity (Horizontal)

import matplotlib.pyplot as plt

languages = ['Python', 'Java', 'C++', 'JavaScript', 'SQL']
values = [95, 80, 70, 85, 75]

plt.figure(figsize=(8,5))

bars = plt.barh(languages, values, color="orange")

plt.title("Programming Languages Popularity")
plt.xlabel("Popularity Score")
plt.ylabel("Languages")

# Add values on bars
for bar in bars:
    width = bar.get_width()
    plt.text(width + 1, bar.get_y() + bar.get_height()/2,
             str(width), va='center')

plt.tight_layout()
plt.savefig("programming_languages_horizontal.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
#E-Commerce Category Revenue

import matplotlib.pyplot as plt

categories = ["Fashion", "Electronics", "Grocery", "Beauty", "Home Decor"]
revenue = [85000, 120000, 65000, 40000, 55000]

plt.figure(figsize=(6,4))
bars = plt.bar(categories, revenue, color="blue")

plt.title("E-Commerce Category Revenue")
plt.xlabel("Categories")
plt.ylabel("Revenue (₹)")

plt.xticks(rotation=20)

# Value labels
for bar in bars:
    y = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, y + 1500, f"{y}", ha='center')

plt.tight_layout()
plt.savefig("ecommerce_category_revenue.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

teams = ['CSK', 'MI', 'RCB', 'KKR', 'SRH']
wins_2024 = [7, 6, 8, 9, 5]
wins_2025 = [8, 7, 9, 6, 7]

x = np.arange(len(teams))
width = 0.35

plt.figure(figsize=(8,5))

bars1=plt.bar(x - width/2, wins_2024, width, label='2024 Season', color='gold')
bars2=plt.bar(x + width/2, wins_2025, width, label='2025 Season', color='red')

# Value labels for 2024 bars
for bar in bars1:
    y = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, y + 0.1, f"{y}", ha='center')

# Value labels for 2025 bars
for bar in bars2:
    y = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, y + 0.1, f"{y}", ha='center')

plt.title("IPL Team Wins Comparison")
plt.xlabel("Teams")
plt.ylabel("Number of Wins")
plt.xticks(x, teams)
plt.legend()

plt.tight_layout()
plt.savefig("ipl_team_wins_comparison.png", dpi=300, bbox_inches='tight')
plt.show()
```
