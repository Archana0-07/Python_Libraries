```python
#Study_hours vs marks in scatter plot using matplotlib

import matplotlib.pyplot as plt

study_hours = [1, 2, 3, 4, 5, 6, 7, 8]
marks = [35, 40, 50, 55, 65, 70, 80, 90]

plt.figure(figsize=(7,5))

plt.scatter(study_hours, marks, s=90)

plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")

plt.grid(True)

plt.savefig("study_hours_vs_marks.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
#Experience vs Salary using matplotlib

import matplotlib.pyplot as plt

experience = [1, 2, 3, 4, 5, 6, 7]
salary = [25000, 30000, 35000, 45000, 52000, 60000, 75000]

plt.figure(figsize=(6,4))

plt.scatter(experience,salary,color="Red",s=80)
plt.title("Experience vs Salary")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.grid(True,linestyle='--')
plt.show()
```

```python
#Age vs Blood Pressure in scatter plot using matplotlib

import matplotlib.pyplot as plt

age = [22, 25, 30, 35, 40, 45, 50, 55]
bp = [110, 115, 118, 122, 130, 135, 140, 145]

plt.figure(figsize=(6,6))
plt.scatter(age,bp,color="black",s=60)
plt.title("Age vs Blood Pressure")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")


plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```
