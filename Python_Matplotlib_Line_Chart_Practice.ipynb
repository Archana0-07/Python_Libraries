```python
#Monthly Sales Trend Analysis using Matplotlib

import matplotlib.pyplot as plt
month=["Jan","Feb","March","April","May","June","July","Aug","Sept","Oct","Nov","Dec"]
sales=[10000,20040,35000,25000,41000,37400,42010,50000,40700,18000,60000,59000]

plt.figure(figsize=(10,5))
plt.plot(month,sales,marker='o', linewidth=3,linestyle="dashed",color="red")

for i in range(len(month)):
    plt.text(month[i], sales[i]+1000, str(sales[i]), ha='center')

plt.savefig("monthly_sales_trend.png", dpi=300, bbox_inches='tight')    
plt.title("Monthly Sales Trend Analysis")
plt.xlabel("Month")
plt.ylabel(" Monthly Sales")
plt.grid(True)
plt.show()
```

```python
#Website Visitors Trend for 7 Days
import matplotlib.pyplot as plt

days=["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
visitors=[101,155,95,140,130,180,210]

plt.figure(figsize=(10,5))
plt.plot(days,visitors,marker='o',linewidth=2,linestyle="dotted",color="blue")

for i in range(len(days)):
    plt.text(days[i], visitors[i]+3, str(visitors[i]),ha="center")
    
plt.savefig("website_visitors_trend.png", dpi=300, bbox_inches='tight')    
plt.title("Website Traffic Trend Visualization")
plt.xlabel("Days")
plt.ylabel("Number of Visitors")
plt.show()
```

```python
#Store Revenue Comparison using Line Chart

import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]

store_a = [10000, 12000, 15000, 17000, 16000, 19000]
store_b = [9000, 11000, 14000, 13000, 18000, 17500]

plt.figure(figsize=(10,5))

plt.plot(months, store_a, marker='o', linewidth=3, label="Store A")
plt.plot(months, store_b, marker='s', linewidth=3, label="Store B")

# Show values on points
for i in range(len(months)):
    plt.text(months[i], store_a[i]+400, str(store_a[i]), ha='center')
    plt.text(months[i], store_b[i]-900, str(store_b[i]), ha='center')

plt.savefig("store_revenue_comparison.png", dpi=300, bbox_inches='tight')
plt.title("Revenue Comparison of Two Stores")
plt.xlabel("Months")
plt.ylabel("Revenue Amount")
plt.grid(True)
plt.legend()
plt.show()
```

```python
#Student Performance Growth Analysis
import matplotlib.pyplot as plt

tests = ["Test 1", "Test 2", "Test 3", "Test 4", "Test 5"]
marks = [65, 72, 80, 78, 90]

plt.figure(figsize=(10,5))

plt.plot(tests, marks, marker='o', linewidth=3, label="Student Marks",linestyle="-.")

# Show marks on points
for i in range(len(tests)):
    plt.text(tests[i], marks[i]+2, str(marks[i]), ha='center')

# Highlight highest mark
max_marks = max(marks)
max_index = marks.index(max_marks)

plt.scatter(tests[max_index], max_marks, s=220, marker='*', label="Highest Mark")

plt.savefig("student_marks_progress.png", dpi=300, bbox_inches='tight')
plt.title("Student Marks Progress in 5 Tests", fontsize=14)
plt.xlabel("Tests")
plt.ylabel("Marks")
plt.ylim(0,100)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```
